#### Basic infos

```python
dataset=pd.read_csv('../anuran_calls/Frogs_MFCCs.csv')
print("Size of the dataset : ",np.shape(dataset))
print("Number of features : ",np.shape(dataset.iloc[:,:-4])[1])
features=list(dataset.columns[:-4])
print("Features:",'\n')
[print(feature) for feature in features]
print('\n')
print("Number of samples : ",np.shape(dataset)[0])
```

#### Distributions of the sample

```python
def CoefficientDistribution(name,values):
    plt.figure()
    plt.hist(values,bins=60)
    plt.title(name+"distribution")
    plt.show()
    pass

for feature in features:
    CoefficientDistribution(feature,dataset[feature])
```

#### Correlation between features

```python
plt.matshow(dataset.corr())
plt.show()
```
